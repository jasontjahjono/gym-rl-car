```python
environment_name = "CarRacing-v2"
env = gym.make(environment_name, render_mode="human")
```

```python
model_path = os.path.join("Training", "Saved Models", "PPO_Driving_Model_8,8m")
```

```python
new_hyperparams = {"learning_rate":0.0001, "gamma":0.99, "batch_size":512, "ent_coef":0.07}
model = PPO(policy='CnnPolicy', env=env, **new_hyperparams, device='cuda', verbose=1, tensorboard_log=log_path)
```

```python
loaded_model = PPO.load(model_path, env=env)
```

```python
copy_params = loaded_model.get_parameters()
copy_params.pop("policy.optimizer", None)
```

```python
model.set_parameters(copy_params, exact_match=False, device='cuda')
```

```python
model.get_parameters()
```

```python
print(model.policy_kwargs)
print("gamma:", model.policy_kwargs["gamma"])
print("ent_coef:", model.policy_kwargs["ent_coef"])
```

```python
model.set_hyperparameters({"learning_rate":0.0002, "gamma":0.95, "batch_size":512, "ent_coef":0.03})
```

```python
model.get_parameters()["policy.optimizer"].keys()
```

```python
environment_name = "CarRacing-v2"
env = gym.make(environment_name, render_mode="human")
obs = env.reset()[0]
# print(obs)
while not (dones and trunc):
#     print(model.predict(obs))
    action, _states = model.predict(obs)
    obs, rewards, dones, trunc, info = env.step(action)
    env.render()
```
